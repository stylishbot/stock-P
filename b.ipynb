```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline

days=20
ticker={
    'apple':'AAPL',
    'facebook':'FB',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'amazon' : 'AMZN',
    'netflix' : 'NFLX'
}

def get_data(days,ticker):
    df=pd.DataFrame()
    for company in ticker.keys():
        tkr=yf.Ticker(ticker[company])
        hist=tkr.history(period=f'{days}d')
        hist.index=hist.index.strftime('%d %B %Y')
        hist=hist[['Close']]
        hist.columns=[company]
        hist=hist.T
        hist.index.name='Name'
        df=pd.concat([df,hist])
    
    return df

```

```python
days=20
ticker={
    'apple':'AAPL',
    'facebook':'FB',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'amazon' : 'AMZN',
    'netflix' : 'NFLX'
}

df=get_data(days,ticker)
```

```python
companies=['apple','facebook','google','microsoft','amazon','netflix']
data=df.loc[companies]
data
```

```python
data=data.T.reset_index()
data.head()
```

```python
data=pd.melt(data, id_vars=['Date']).rename(
    columns={'value':'Stock Prices(USD)'}
)
data
```

```python
chart=(
    alt.Chart(data)
    .mark_line(opacity=0.8, clip=True)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD)",stack=None,scale=alt.Scale(domain=[0,300])),
        color='Name:N'
    )
)
```
